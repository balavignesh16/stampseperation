```python
# Google Colab Notebook for Training Mask R-CNN on Stamps & Signatures Dataset

# Install Dependencies
!pip install torch torchvision torchaudio
!pip install detectron2 -f https://dl.fbaipublicfiles.com/detectron2/wheels/cu118/torch2.0/index.html

# Import Required Libraries
import torch
import detectron2
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2.data import MetadataCatalog, DatasetCatalog
from detectron2.data.datasets import register_coco_instances
import os

# Mount Google Drive for Dataset Storage
from google.colab import drive
drive.mount('/content/drive')

# Define Dataset Path
DATASET_PATH = "/content/drive/MyDrive/stamp_signature_dataset/"
ANNOTATION_PATH = "/content/drive/MyDrive/stamp_signature_dataset/VIT_Dataset/images/annotation.json"
IMAGE_PATH = "/content/drive/MyDrive/stamp_signature_dataset/VIT_Dataset/"

# Register Custom Dataset
register_coco_instances("stamps_signatures", {}, ANNOTATION_PATH, IMAGE_PATH)
metadata = MetadataCatalog.get("stamps_signatures")
dataset_dicts = DatasetCatalog.get("stamps_signatures")

# Configure Mask R-CNN Model
cfg = get_cfg()
cfg.merge_from_file("detectron2/configs/COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")
cfg.DATASETS.TRAIN = ("stamps_signatures",)
cfg.DATASETS.TEST = ()
cfg.DATALOADER.NUM_WORKERS = 2
cfg.MODEL.WEIGHTS = "detectron2://COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x/137849600/model_final_f10217.pkl"
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 1000
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2  # Stamps and Signatures

# Train the Model
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()

# Save the Trained Model
model_output_path = "/content/drive/MyDrive/mask_rcnn_stamps_signatures.pth"
torch.save(trainer.model.state_dict(), model_output_path)

print(f"Model saved at {model_output_path}")

# Inference and Testing
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer
import cv2

cfg.MODEL.WEIGHTS = model_output_path  # Load Trained Weights
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7
predictor = DefaultPredictor(cfg)

# Load a Test Image
test_image_path = os.path.join(IMAGE_PATH, "test_sample.jpg")
image = cv2.imread(test_image_path)
outputs = predictor(image)

# Visualize Predictions
visualizer = Visualizer(image[:, :, ::-1], metadata=metadata)
out = visualizer.draw_instance_predictions(outputs["instances"].to("cpu"))
cv2.imshow("Predictions", out.get_image()[:, :, ::-1])
```

```python
import os
import json

# Define paths
DATASET_NAME = "stamps"
IMAGE_DIR = "/content/drive/MyDrive/stamp_signature_dataset/images"
ANNOTATION_FILE = "/content/drive/MyDrive/stamp_signature_dataset/annotations.json"

# Check if files exist
if os.path.exists(ANNOTATION_FILE) and os.path.exists(IMAGE_DIR):
    print("✅ Dataset paths verified!")
else:
    print("❌ Check dataset paths! Some files are missing.")
```

```python
from pycocotools.coco import COCO

# Load COCO annotations
coco = COCO(ANNOTATION_FILE)

# Get all image IDs
image_ids = coco.getImgIds()
print(f"✅ Found {len(image_ids)} images in dataset.")

# Get all categories
categories = coco.loadCats(coco.getCatIds())
category_names = [cat["name"] for cat in categories]
print(f"✅ Categories: {category_names}")
```

```python
import cv2
import random
import matplotlib.pyplot as plt
from pycocotools.coco import COCO

# Function to visualize an image with annotations
def visualize_sample():
    img_id = random.choice(image_ids)  # Select a random image ID
    img_info = coco.loadImgs(img_id)[0]
    img_path = os.path.join(IMAGE_DIR, img_info["file_name"])

    # Load image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Get annotations
    ann_ids = coco.getAnnIds(imgIds=img_id)
    anns = coco.loadAnns(ann_ids)

    # Draw bounding boxes
    for ann in anns:
        x, y, w, h = ann["bbox"]
        cv2.rectangle(img, (int(x), int(y)), (int(x + w), int(y + h)), (255, 0, 0), 2)

    # Display image
    plt.figure(figsize=(8, 6))
    plt.imshow(img)
    plt.axis("off")
    plt.show()

# Run visualization
visualize_sample()
```

```python
import tensorflow as tf

# Function to parse COCO annotations into TensorFlow format
def parse_coco_dataset():
    dataset = []
    for img_id in image_ids:
        img_info = coco.loadImgs(img_id)[0]
        ann_ids = coco.getAnnIds(imgIds=img_id)
        anns = coco.loadAnns(ann_ids)

        # Store annotations
        annotations = []
        for ann in anns:
            x, y, w, h = ann["bbox"]
            annotations.append({"bbox": (x, y, w, h), "category_id": ann["category_id"]})

        dataset.append({
            "image_path": os.path.join(IMAGE_DIR, img_info["file_name"]),
            "annotations": annotations
        })

    return dataset

# Load dataset
tf_dataset = parse_coco_dataset()
print(f"✅ Loaded {len(tf_dataset)} images for training!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
image_path = "/content/drive/MyDrive/stamp_signature_dataset/images"
# Load images and labels
def load_data():
    images, labels = [], []
    for data in tf_dataset:
        # Load image
        img = load_img(data["image_path"], target_size=(224, 224))  # Resize if needed
        img = img_to_array(img) / 255.0  # Normalize

        # Extract bounding boxes (for simplicity, using only first bbox per image)
        if data["annotations"]:
            bbox = data["annotations"][0]["bbox"]  # (x, y, w, h)
            label = data["annotations"][0]["category_id"]
        else:
            bbox = (0, 0, 0, 0)  # No bbox
            label = 0

        images.append(img)
        labels.append(bbox)  # You can modify this to include class labels

    return tf.convert_to_tensor(images), tf.convert_to_tensor(labels)

# Load dataset
X_train, y_train = load_data()
print(f"✅ Training data shape: {X_train.shape}, Labels shape: {y_train.shape}")

# Build a simple CNN Model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation="relu", input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(4)  # Output = 4 (x, y, width, height)
])

# Compile model
model.compile(optimizer="adam", loss="mse", metrics=["mae"])

# Train model
model.fit(X_train, y_train, epochs=10, batch_size=8)
```

```python
import torch
import torchvision
import os
import numpy as np
import torchvision.transforms as T
from torchvision.models.detection import maskrcnn_resnet50_fpn
from pycocotools.coco import COCO
from PIL import Image

# Custom COCO Dataset class
class CustomCocoDataset(torch.utils.data.Dataset):
    def __init__(self, root, annotation_file, transforms=None):
        self.root = root
        self.coco = COCO(annotation_file)
        self.ids = list(self.coco.imgs.keys())
        self.transforms = transforms

    def __getitem__(self, index):
        img_id = self.ids[index]
        annotation = self.coco.loadAnns(self.coco.getAnnIds(imgIds=img_id))

        # Load image
        img_path = os.path.join(self.root, self.coco.loadImgs(img_id)[0]['file_name'])
        image = Image.open(img_path).convert("RGB")

        # Convert annotations to PyTorch format
        masks = []
        boxes = []
        labels = []

        for ann in annotation:
            masks.append(self.coco.annToMask(ann))
            x, y, w, h = ann["bbox"]
            boxes.append([x, y, x + w, y + h])
            labels.append(ann["category_id"])

        # Convert to tensors
        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        masks = torch.as_tensor(np.array(masks, dtype=np.uint8), dtype=torch.uint8)

        target = {"boxes": boxes, "labels": labels, "masks": masks}

        if self.transforms:
            image = self.transforms(image)

        return image, target

    def __len__(self):
        return len(self.ids)

# Set up transform and dataset
transform = T.Compose([T.ToTensor()])
dataset = CustomCocoDataset(root="/content/drive/MyDrive/stamp_signature_dataset/images",
                            annotation_file="/content/drive/MyDrive/stamp_signature_dataset/annotations.json",
                            transforms=transform)

# DataLoader
data_loader = torch.utils.data.DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))

# Device setup
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# Initialize Mask R-CNN model
model = maskrcnn_resnet50_fpn(pretrained=False, num_classes=len(dataset.coco.getCatIds()) + 1)  # +1 for background
model.to(device)

# Optimizer and criterion
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
criterion = torch.nn.CrossEntropyLoss()

# Training Loop
num_epochs = 10  # Adjust as needed

for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for images, targets in data_loader:
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        loss = sum(loss for loss in loss_dict.values())

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss/len(data_loader)}")

# Save model
torch.save(model.state_dict(), "/content/drive/MyDrive/mask_rcnn_stamp_removal.pth")

# Inference on a test image
model.eval()
image = Image.open("/content/drive/MyDrive/stamp_signature_dataset/images/test_image.jpg").convert("RGB")
image = transform(image).to(device)

with torch.no_grad():
    prediction = model([image])

# Visualizing the result
import matplotlib.pyplot as plt

plt.imshow(image.cpu().permute(1, 2, 0))
for mask in prediction[0]["masks"]:
    plt.imshow(mask[0].cpu(), alpha=0.5)
plt.show()
```

```python
import torchvision
from torchvision.models.detection import MaskRCNN
from torchvision.models.detection.rpn import AnchorGenerator

# Load a base ResNet50 backbone and initialize Mask R-CNN
backbone = torchvision.models.resnet50(pretrained=False)
backbone.out_channels = 2048  # Adjust output channels

# Define anchor generator
anchor_generator = AnchorGenerator(sizes=((32, 64, 128, 256, 512),),
                                   aspect_ratios=((0.5, 1.0, 2.0),) * 5)

# Define ROI Pooler
roi_pooler = torchvision.ops.MultiScaleRoIAlign(featmap_names=['0'], output_size=7, sampling_ratio=2)

# Create Mask R-CNN model
model = MaskRCNN(backbone, num_classes=2, rpn_anchor_generator=anchor_generator, box_roi_pool=roi_pooler)

# Move model to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
import torch
import torch.optim as optim

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Training Loop
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for images, targets in data_loader:
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())
        losses.backward()
        optimizer.step()

        total_loss += losses.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}")
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from torchvision.models.detection.mask_rcnn import MaskRCNNPredictor
from torchvision.transforms import functional as F
from PIL import Image
import os
import json

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Mask R-CNN model
model = torchvision.models.detection.maskrcnn_resnet50_fpn(pretrained=True)
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes=2)  # 2 classes: background & stamp

# Modify mask predictor
in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
hidden_layer = 256
model.roi_heads.mask_predictor = MaskRCNNPredictor(in_features_mask, hidden_layer, num_classes=2)

model.to(device)

# Define dataset class
class StampDataset(torch.utils.data.Dataset):
    def __init__(self, image_dir, annotation_path, transform=None):
        self.image_dir = image_dir
        self.annotation_path = annotation_path
        self.transform = transform

        # Load COCO-style annotations
        with open(annotation_path) as f:
            self.annotations = json.load(f)

        self.image_ids = list(set(ann["image_id"] for ann in self.annotations["annotations"]))

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        image_info = next(img for img in self.annotations["images"] if img["id"] == image_id)
        image_path = os.path.join(self.image_dir, image_info["file_name"])

        image = Image.open(image_path).convert("RGB")
        width, height = image.size

        # Get annotations
        boxes, masks, labels = [], [], []
        for ann in self.annotations["annotations"]:
            if ann["image_id"] == image_id:
                x, y, w, h = ann["bbox"]
                boxes.append([x, y, x + w, y + h])
                labels.append(1)  # 1 = stamp
                mask = Image.new("L", (width, height), 0)
                for segment in ann["segmentation"]:
                    mask.paste(255, [int(p) for p in segment])
                masks.append(F.to_tensor(mask).squeeze())

        # Convert to tensors
        image = F.to_tensor(image)
        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)
        masks = torch.stack(masks) if masks else torch.zeros((0, height, width), dtype=torch.uint8)

        target = {"boxes": boxes, "labels": labels, "masks": masks}

        return image, target

# Load dataset
image_dir = "/content/drive/MyDrive/stamp_signature_dataset/images/"
annotation_path = "/content/drive/MyDrive/stamp_signature_dataset/annotations.json"

dataset = StampDataset(image_dir, annotation_path)
data_loader = DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))

# Define optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0

    for images, targets in data_loader:
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        loss = sum(loss for loss in loss_dict.values())  # Compute total loss
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"📉 Epoch {epoch+1}: Loss = {epoch_loss:.4f}")

# Save trained model
torch.save(model.state_dict(), "mask_rcnn_stamp_detector.pth")
print("✅ Model training complete and saved!")
```

```python
import torch
import torchvision.transforms.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image, ImageDraw
import json
import numpy as np
import os

class StampDataset(Dataset):
    def __init__(self, image_dir, annotation_file, transforms=None):
        self.image_dir = image_dir
        self.transforms = transforms

        # Load COCO annotations
        with open(annotation_file, "r") as f:
            self.coco_data = json.load(f)

        self.images = {img["id"]: img for img in self.coco_data["images"]}
        self.annotations = {ann["image_id"]: [] for ann in self.coco_data["annotations"]}

        for ann in self.coco_data["annotations"]:
            self.annotations[ann["image_id"]].append(ann)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_info = list(self.images.values())[idx]
        img_path = os.path.join(self.image_dir, img_info["file_name"])
        image = Image.open(img_path).convert("RGB")
        width, height = image.size

        annots = self.annotations.get(img_info["id"], [])
        boxes = []
        masks = []
        labels = []

        for ann in annots:
            bbox = ann["bbox"]
            x, y, w, h = bbox
            boxes.append([x, y, x + w, y + h])
            labels.append(1)  # Single class: "stamp"

            # Create binary mask
            mask = Image.new("L", (width, height), 0)
            segmentation = np.array(ann["segmentation"][0]).reshape(-1, 2)
            ImageDraw.Draw(mask).polygon(list(segmentation.flatten()), outline=1, fill=1)
            masks.append(F.to_tensor(mask).squeeze())

        # Convert to tensors
        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        masks = torch.stack(masks) if masks else torch.zeros((0, height, width))

        target = {"boxes": boxes, "labels": labels, "masks": masks}

        if self.transforms:
            image = self.transforms(image)

        return image, target
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from pycocotools.coco import COCO
from PIL import Image
import matplotlib.pyplot as plt

# ------------------------------
# 1. Custom Collate Function
# ------------------------------
def collate_fn(batch):
    # Returns (list of images, list of targets)
    return tuple(zip(*batch))

# ------------------------------
# 2. Custom Dataset Class
# ------------------------------
class CustomDataset(Dataset):
    def __init__(self, image_dir, annotation_file, transform=None):
        self.image_dir = image_dir  # Base directory for images
        self.annotation_file = annotation_file
        self.transform = transform
        self.coco = COCO(self.annotation_file)
        self.image_ids = list(self.coco.imgs.keys())

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        image_info = self.coco.imgs[image_id]

        # Build full image path using the base directory and the file_name from annotations.
        # If the annotation's file_name is "../Декларация-соответствия-1 (1).jpg",
        # then this will resolve to:
        # "/content/drive/MyDrive/stamp_signature_dataset/images/../Декларация-соответствия-1 (1).jpg"
        # which normalizes to "/content/drive/MyDrive/stamp_signature_dataset/Декларация-соответствия-1 (1).jpg".
        # If your images are stored inside "images", update your annotations to remove the "../".
        img_path = os.path.normpath(os.path.join(self.image_dir, image_info['file_name']))
        # Uncomment for debugging:
        # print("Loading image from:", img_path)
        image = Image.open(img_path).convert("RGB")

        # Load annotations for this image
        ann_ids = self.coco.getAnnIds(imgIds=[image_id])
        annotations = self.coco.loadAnns(ann_ids)

        boxes = []
        labels = []
        for ann in annotations:
            x, y, w, h = ann['bbox']
            if w > 0 and h > 0:  # Only include valid boxes
                # Convert COCO bbox [x, y, w, h] to [xmin, ymin, xmax, ymax]
                boxes.append([x, y, x + w, y + h])
                labels.append(1)  # All annotations assumed to belong to class "stamp/signature"

        # If no valid boxes are found, skip this image by returning the next one
        if len(boxes) == 0:
            return self.__getitem__((idx + 1) % len(self))

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)
        target = {'boxes': boxes, 'labels': labels}

        if self.transform:
            image = self.transform(image)

        return image, target

# ------------------------------
# 3. Model Setup Function
# ------------------------------
def get_model(num_classes):
    model = fasterrcnn_resnet50_fpn(pretrained=True)
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor.cls_score = torch.nn.Linear(in_features, num_classes)
    model.roi_heads.box_predictor.bbox_pred = torch.nn.Linear(in_features, num_classes * 4)
    return model

# ------------------------------
# 4. Paths, Transforms, and DataLoader
# ------------------------------
# Set the base image directory to the images folder.
img_dir = "/content/drive/MyDrive/stamp_signature_dataset/images"
annotation_file = "/content/drive/MyDrive/stamp_signature_dataset/annotations.json"

transform = transforms.Compose([
    transforms.Resize((800, 800)),
    transforms.ToTensor(),
])

dataset = CustomDataset(image_dir=img_dir, annotation_file=annotation_file, transform=transform)
data_loader = DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=collate_fn)

# ------------------------------
# 5. Model, Optimizer, and Device Setup
# ------------------------------
num_classes = 2  # 1 for stamp/signature + 1 for background
model = get_model(num_classes)
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)

params = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.SGD(params, lr=0.005, momentum=0.9, weight_decay=0.0005)

# ------------------------------
# 6. Training Loop
# ------------------------------
num_epochs = 10  # Adjust as needed
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0
    for images, targets in data_loader:
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())
        epoch_loss += losses.item()
        losses.backward()
        optimizer.step()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")

# ------------------------------
# 7. Save the Model
# ------------------------------
torch.save(model.state_dict(), "faster_rcnn_model.pth")

# ------------------------------
# 8. Evaluation and Visualization (Optional)
# ------------------------------
model.eval()
sample_image, sample_target = dataset[0]
with torch.no_grad():
    prediction = model([sample_image.to(device)])

image_np = sample_image.permute(1, 2, 0).cpu().numpy()
boxes = prediction[0]['boxes'].cpu().numpy()

fig, ax = plt.subplots(1)
ax.imshow(image_np)
for box in boxes:
    rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                         fill=False, color='r', linewidth=3)
    ax.add_patch(rect)
plt.show()
```
